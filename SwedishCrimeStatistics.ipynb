# Swedish Crime Statistics

** Статистика преступлений в Швеции в 1950-2015 годах **

Набор данных содержит следующие столбцы:
- `crimes.total` - Общее количество зарегистрированных преступлений
- `crimes.penal.code` - Общее количество зарегистрированных уголовных преступлений
- `crimes.person` - Количество зарегистрированных преступлений, совершенных против личности
- `murder` - Количество убийств
- `sexual.offences` - Количество зарегистрированных преступлений, совершенных на сексуальной почве
- `rape` - Количество зарегистрированных изнасилований
- `assault` - Количество зарегистрированных нападений
- `stealing.general` - Общее количество зарегистрированных преступлений, связанных с кражами/грабежами
- `robbery` - Количество зарегистрированных ограблений
- `burglary` - Количество зарегистрированных краж со взломом
- `vehicle.theft` - Количество зарегистрированных угонов автомобилей
- `house.theft` - Количество зарегистрированных домашних краж
- `shop.theft` - Количество зарегистрированных магазинных краж
- `out.of.vehicle.theft` - Количество зарегистрированных краж из автомобилей
- `criminal.damage` - Количество зарегистрированных причинений ущерба
- `other.penal.crimes` - Другие зарегистрированные уголовные преступления
- `fraud` - Количество зарегистрированных мошенничеств
- `narcotics` - Количество зарегистрированных злоупотреблений наркотиками
- `drunk.driving` - Количество зарегистрированных инцидентов, связанных с вождением в нетрезвом виде
- `Year` - Год
- `population` - Население

# Задача

**Ставим задачу определения уровня преступности.**

Введем **категориальный показатель уровня преступности**:
* < 50 преступлений на 100 000 человек населения - Низкий уровень преступности (**1**)
* < 100 преcтуплений на 100 000 человек населения - Ниже среднего уровень преступности (**2**)
* < 150 преступлений на 100 000 человек населения - Средний уровень преступности (**3**)
* < 200 преступлений на 100 000 человек населения - Выше среднего уровень преступности (**4**)
* \> 200 преступлений на 100 000 человек населения - Высокий уровень преступности (**5**) 

## Подготовка данных


### Обработка пропущенных значений

```python
data.count(axis = 0)
```

```python
empty_columns = [c for c in data.columns if data[c].count() < 66]
print(empty_columns)
```

```python
data[empty_columns].describe()
```

```python
pd.concat((data[empty_columns], data['crimes.total']), axis = 1).corr()
```

Пропущенные значения имются в пяти столбцах таблицы.

Так как среднеквадратическое отклонение в некоторых столбцах большое, например в `narcotics`, то не можем просто заполнить таблицу средними значениями, так как это даст неправдободобные данные.

Предлагается следующее решение:

Высчитаем зависимость между значениями в этих столбцах и общим числом преступлений и с помощью нее заполним таблицу.

Так как данные изменяются очень сильно от начала к концу таблицы, а пропущенные значения располагаются в начале таблицы, зависимость будем искать в первой половине таблицы.

```python
years_to_calc_median = 25
for c in empty_columns:
    axes_to_join = data.drop(data[data[c].isnull()].index)
    data_without_empty = (pd.concat((axes_to_join[c], axes_to_join['crimes.total']), axis = 1)).head(years_to_calc_median)
    median = (data_without_empty[c]/data_without_empty['crimes.total']).median()
    data[c] = data[c].fillna(round(data['crimes.total'] * median))
```

```python
data[empty_columns].head(15)
```

```python
data.count(axis = 0)
```

### Работа над данными

Заменим абсолютные показатели преступности на относительные.

Заменяем число преступлений на их удельный вес.

```python
crimes_columns = [c for c in data.columns if (c != 'Year') & (c != 'population') & (c != 'crimes.total')]
print(crimes_columns)
```

```python
for c in crimes_columns:
    data[c] = (data[c]/data['crimes.total'])*100
```

Заменяем общее число преступлений на уровень преступности и затем на соответствующее категориальное значение.

Переименовываем `crimes.total` в `crime.level`.

```python
measure_population = 100000
data['crimes.total'] = (data['crimes.total'] / data['population']) * measure_population

data.at[data['crimes.total'] < 50, 'crimes.total'] = 1
data.at[data['crimes.total'] >= 200, 'crimes.total'] = 5
data.at[data['crimes.total'] >= 150, 'crimes.total'] = 4
data.at[data['crimes.total'] >= 100, 'crimes.total'] = 3
data.at[data['crimes.total'] >= 50, 'crimes.total'] = 2

data.rename(columns = {"crimes.total" : "crime.level"}, inplace = True)
data['crime.level']
```

```python
data
```

### Нормализация

Так как все столбцы, за исключением `crime.level`, имеют количественные значения, просто нормализуем их.

```python
test_columns = [c for c in data.columns if c != 'crime.level']
test_data = data[test_columns]
test_data
```

```python
test_data = (test_data - test_data.mean(axis = 0))/test_data.std(axis = 0)
test_data.describe()
```

Объединяем данные в одну таблицу.

```python
data = pd.concat((test_data, data['crime.level']), axis = 1)
data
```

### Разделение на обучающую и тестовую выборки

```python
X = data.drop(('crime.level'), axis = 1)
y = data['crime.level']
feature_names = X.columns
```

```python
print(X.shape)
print(y.shape)
N, d = X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

## $k$NN

Применяем алгоритм машинного обучения **$K$ ближайших соседей** с параметром 10.

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train, y_train)
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)
```

Высчитываем ошибки на обучающей и тестовой выборках соответственно.

```python
err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

```python
print(1 - knn.score(X_test, y_test))
```

Смотрим матрицу ошибок.

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

### Подбор параметров

Протестируем выборки методом **$k$NN** с параметрами от 1 до 25.

```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=9)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

### Метод $k$NN с найденным оптимальным значением параметра.

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))
```

Ошибки на обучающей и тестовой выборках соответственно:

```python
print(err_train, err_test)
```

Матрица ошибок:

```python
print(confusion_matrix(y_test, y_test_predict))
```

Методом **$k$NN** лучшие результаты получаются при рассмотрении трех ближайших соседей.

Можно предположить, что такой результат получается ввиду общего сильного (почти линейного) роста уровня преступности в рассматриваемый период.

## Multi-layer perceptron (MLP)

Применяем для обучения нейронную сеть.

```python
from sklearn.neural_network import MLPClassifier
```

### Подбор количества скрытых слоев (1 или 2)

```python
layers = [1, 2]
test_err = []
train_err = []
for layer in layers:
    mlp_model = MLPClassifier(hidden_layer_sizes = (100, layer), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(layers, train_err, 'b-o', label = 'train')
plt.semilogx(layers, test_err, 'r-o', label = 'test')
plt.title('Number of hidden layers vs. alpha')
plt.xlabel('number of hidden layers')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)

print(min_train_err, min_test_err)
```

Ошибка меньше при использовании нейронной сети с двумя скрытыми слоями.

```python
layer_opt = 2
```

### Подбор *гипер*параметра *alpha*

```python
alpha_arr = np.logspace(-3, 2, 21)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (100, layer_opt), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'train')
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)

print(min_train_err, min_test_err)
```

```python
alpha_opt = alpha_arr[test_err == min_test_err]
print(alpha_opt)
```

```python
alpha_opt = alpha_opt[0]
```

### Подбор количества нейронов в слоях

```python
neuron_arr = [50, 100, 500, 1000, 5000]
test_err = []
train_err = []
for neuron in neuron_arr:
    mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (neuron, layer_opt), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
plt.semilogx(neuron_arr, train_err, 'b-o', label = 'train')
plt.semilogx(neuron_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.min(neuron_arr), np.max(neuron_arr)])
plt.title('Error vs. number of neurons')
plt.xlabel('number of neurons')
plt.ylabel('error')
plt.legend()
pass
```

```python
min_train_err = np.min(train_err)
min_test_err = np.min(test_err)

print(min_train_err, min_test_err)
```

Наилучшие результаты для данной задачи получаются при использовании сравнительно малого количества нейронов в слоях сети.

```python
neuron_opt = 50
```

### Алгорит MLP с подобранными оптимальными параметрами

Количество скрытых слоев = 2

Гиперпараметр alpha = 0.03162278

Количество нейронов в слоях = 50

```python
mlp_model = MLPClassifier(alpha = alpha_opt, hidden_layer_sizes = (neuron_opt, layer_opt), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)

y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)
```

```python
err_train = np.mean(y_train != y_train_pred)
err_test  = np.mean(y_test  != y_test_pred)

print(err_train, err_test)
```

```python
print(1 - mlp_model.score(X_test, y_test))
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_pred))
```

**MLP** дал лучшие результаты по сравнению с **$k$NN**.
